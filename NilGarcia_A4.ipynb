```python

def fibonnacci(n):
    llista = []
    if n==0:
        return [0]
    elif n==1:
        return [0, 1]
    
    llista = [0,1]
    
    for _ in range(2, n):
        llista.append(llista[-1]+llista[-2])
    return llista
            


fibonnacci(99)
    
```

```python
def calcularpoblacio(P0,r):
    return P0*(1+r)

def creixementpoblacio(P0, r, d):
    poblacions = [P0]
    for i in range(d):
        P0 = calcularpoblacio(P0, r)
        poblacions.append(P0)
    return poblacions


final = creixementpoblacio(1000, 0.01, 365)
print(final[-1])
```

```python
from IPython.display import clear_output
N = int(input("Introdueix la mida del taulell: ") or 3)
def mostra_taulell(taulell, diccionari_jugadors):
    print("    " + "   ".join(str(i) for i in range(N)))
    for fila in range(N):
        contingut = " | ".join(diccionari_jugadors[taulell[fila][columna]] for columna in range(N))
        print(f" {fila}  {contingut}")
        if fila < N-1:
            print("   " + "---+" *(N-1) + "---")
    print()
    return


def demana_posicio():
    try:
        fila = int(input(f"Dona'm la fila on vols posar la peça (valor entre 0 i {N-1})"))
        while fila < 0 or fila > N-1:
            fila = int(input(f"Siusplau, dona'm be la fila on vols posar la peça (valor entre 0 i {N-1})"))
    except:
        return None, None
    try:
        columna = int(input(f"Dona'm la columna on vols posar la peça (valor entre 0 i {N-1})"))
        while columna < 0 or columna > N-1:
            columna = int(input(f"Siusplau, dona'm be la columna on vols posar la peça (valor entre 0 i {N-1})"))
    except:
        return None, None
    if taulell[fila][columna] != 0:
        print("No pots posar la peça aquí maquina, try again.")
        return demana_posicio()
    return fila, columna

def actualitza_taulell(taulell, fila, columna, jugador):
    taulell[fila][columna] = jugador
    return taulell

def canvi_jugador(jugador):
    return jugador*(-1)

def comprovar_victoria(taulell,jugador):
    # Victòria per files
    for fila in range(0,N):
        suma = 0
        for columna in range(0,N):
            suma += taulell[fila][columna]
        if suma == N*jugador:
            return True
    # Victòria per columnes
    for columna in range(0,N):
        suma = 0
        for fila in range(0,N):
            suma += taulell[fila][columna]

        if suma == N*jugador:
            return True 
    # Victòria per diagonals
    suma = 0
    for diagonal in range(0,N):
        suma += taulell[diagonal][diagonal]            
    if suma == N*jugador:
        return True 
    # Victòria per diagonal
    suma = 0
    for diagonal in range(0,N):
        suma += taulell[diagonal][N-1-diagonal]            
    if suma == N*jugador:
        return True 
    
    return False

jugador = 1
diccionari_jugadors = {
    0: " ",
    1: "X",
    -1: "O"
    }

taulell=[[0]*N for _ in range(N)]
condicio_victoria = False
comptador = 0

while not condicio_victoria and comptador < N**2:
    clear_output(wait=True)
    mostra_taulell(taulell,diccionari_jugadors)
    fila, columna = demana_posicio()
    if fila == None and columna == None:
        print(f"Has de donar un nombre natural entre 0 i {N-1}. Sortim del joc")
        break
    taulell = actualitza_taulell(taulell, fila, columna, jugador)
    condicio_victoria = comprovar_victoria(taulell, jugador)
    if condicio_victoria:
        clear_output(wait=True)
        mostra_taulell(taulell,diccionari_jugadors)
        print(f"Ha guanyat el jugador {diccionari_jugadors[jugador]}")
        break

    comptador += 1
    jugador = canvi_jugador(jugador)
```
